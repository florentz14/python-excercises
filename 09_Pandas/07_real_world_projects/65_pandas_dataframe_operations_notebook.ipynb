# Pandas DataFrame Operations Notebook

Complete practice notebook for core Pandas DataFrame operations.

```python
import pandas as pd
import numpy as np
from pathlib import Path

print("=" * 65)
print("   PANDAS DATAFRAME OPERATIONS")
print("=" * 65)
```

## 1) DataFrame creation from dictionary

```python
data = {
    "nombre": ["Ana", "Luis", "Maria", "Carlos", "Sofia"],
    "edad": [23, 30, 27, 35, 22],
    "ciudad": ["Madrid", "Barcelona", "Madrid", "Barcelona", "Sevilla"],
}
df = pd.DataFrame(data)
df
```

## 2) DataFrame from list of dictionaries

```python
list_of_dicts = [
    {"nombre": "Pedro", "edad": 28, "ciudad": "Valencia"},
    {"nombre": "Laura", "edad": 33, "ciudad": "Bilbao"},
    {"nombre": "Tomas", "edad": 25, "ciudad": "Madrid"},
]
df2 = pd.DataFrame(list_of_dicts)
df2
```

## 3) DataFrame from NumPy array

```python
np.random.seed(42)
df3 = pd.DataFrame(np.random.randint(1, 100, size=(4, 3)), columns=["A", "B", "C"])
df3
```

## 4) Load CSV data with a self-contained sample

```python
csv_content = """producto,ventas,gastos,region
Laptop,15000,8000,Norte
Tablet,9000,4500,Sur
Monitor,7200,3100,Norte
Teclado,3400,1200,Este
Raton,2100,900,Sur
Impresora,5600,2800,Oeste
Auriculares,4300,1900,Norte
Webcam,3800,1600,Este
Laptop,17000,9500,Sur
Tablet,11000,5200,Oeste
"""

temp_csv = Path("ventas_temp.csv")
temp_csv.write_text(csv_content, encoding="utf-8")

df4 = pd.read_csv(temp_csv)
df4.head()
```

## 5) Selection, indexing and filtering

```python
print("Single column:")
print(df["nombre"].to_frame())

print("\nMultiple columns:")
print(df[["nombre", "edad"]])

print("\nRow selection with iloc:")
print(df.iloc[0])
print(df.iloc[1:3])

print("\nRow selection with loc:")
print(df.loc[0])

print("\nFilter edad > 25:")
print(df[df["edad"] > 25])

print("\nFilter edad > 25 AND ciudad == 'Barcelona':")
print(df[(df["edad"] > 25) & (df["ciudad"] == "Barcelona")])
```

## 6) Add, calculate and remove columns/rows

```python
df["salario"] = [28000, 45000, 33000, 60000, 25000]
df["salario_mensual"] = df["salario"] / 12

print("With new columns:")
print(df)

df = df.drop("salario_mensual", axis=1)
print("\nAfter dropping salario_mensual:")
print(df)

df_sin_primera = df.drop(0, axis=0)
print("\nAfter dropping first row:")
print(df_sin_primera)
```

## 7) Missing values operations

```python
datos_faltantes = {
    "nombre": ["Ana", "Luis", "Maria", "Carlos", "Sofia"],
    "edad": [23, None, 27, None, 22],
    "salario": [28000, 45000, None, 60000, None],
}
df_nan = pd.DataFrame(datos_faltantes)

print("DataFrame with NaN values:")
print(df_nan)

print("\nNull matrix:")
print(df_nan.isnull())

print("\nNull count by column:")
print(df_nan.isnull().sum())

df_dropped = df_nan.dropna()
print("\nAfter dropna():")
print(df_dropped)

df_filled_zero = df_nan.fillna(0)
print("\nAfter fillna(0):")
print(df_filled_zero)

df_filled_mean = df_nan.copy()
df_filled_mean["edad"] = df_filled_mean["edad"].fillna(df_filled_mean["edad"].mean())
df_filled_mean["salario"] = df_filled_mean["salario"].fillna(df_filled_mean["salario"].mean())

print("\nAfter fillna(mean):")
print(df_filled_mean)
```

## 8) Practical case: profit, groupby and maxima

```python
df4["beneficio"] = df4["ventas"] - df4["gastos"]
print("Data with beneficio:")
print(df4)

por_region = pd.DataFrame(
    df4.groupby("region", as_index=False).agg(
        ventas=("ventas", "sum"),
        gastos=("gastos", "sum"),
        beneficio=("beneficio", "sum"),
    )
)
print("\nGroup by region:")
print(por_region)

por_producto = pd.DataFrame(
    df4.groupby("producto", as_index=False).agg(
        ventas=("ventas", "sum"),
        gastos=("gastos", "sum"),
        beneficio=("beneficio", "sum"),
    )
)
print("\nGroup by product:")
print(por_producto)

por_region_sorted = pd.DataFrame(
    sorted(por_region.itertuples(index=False), key=lambda row: float(row[3]), reverse=True),
    columns=["region", "ventas", "gastos", "beneficio"],
)
por_producto_sorted = pd.DataFrame(
    sorted(por_producto.itertuples(index=False), key=lambda row: float(row[3]), reverse=True),
    columns=["producto", "ventas", "gastos", "beneficio"],
)

region_max = str(por_region_sorted.iloc[0]["region"])
producto_max = str(por_producto_sorted.iloc[0]["producto"])

print("\nRegion with highest beneficio:", region_max)
print("Beneficio:", float(por_region_sorted.iloc[0]["beneficio"]))

print("\nProduct with highest beneficio:", producto_max)
print("Beneficio:", float(por_producto_sorted.iloc[0]["beneficio"]))
```

## 9) Operation summary

```python
operations = [
    ("1", "DataFrame from dictionary", "pd.DataFrame(dict)"),
    ("2", "DataFrame from list of dicts", "pd.DataFrame([{...}])"),
    ("3", "DataFrame from NumPy array", "pd.DataFrame(np.array, columns=[...])"),
    ("4", "Load CSV", "pd.read_csv('file.csv')"),
    ("5", "Column selection", "df['col'] / df[['c1','c2']]"),
    ("6", "Row selection by position", "df.iloc[0] / df.iloc[1:3]"),
    ("7", "Row selection by label", "df.loc[0]"),
    ("8", "Conditional filtering", "df[df['col'] > value]"),
    ("9", "Multiple condition filtering", "df[(cond1) & (cond2)]"),
    ("10", "Add column", "df['new'] = [...]"),
    ("11", "Calculated column", "df['col'] = df['other'] / 12"),
    ("12", "Drop column", "df.drop('col', axis=1)"),
    ("13", "Drop row", "df.drop(0, axis=0)"),
    ("14", "DataFrame with NaN", "pd.DataFrame(data_with_none)"),
    ("15", "Null detection", "isnull() / isnull().sum()"),
    ("16", "Drop null rows", "dropna()"),
    ("17", "Fill nulls with zero", "fillna(0)"),
    ("18", "Fill nulls with mean", "fillna(df['col'].mean())"),
    ("19", "Profit calculation", "df['profit'] = sales - costs"),
    ("20", "Group by region", "groupby('region').sum()"),
    ("21", "Group by product", "groupby('product').sum()"),
    ("22", "Find maximum with idxmax", "groupby(...).sum().idxmax()"),
]

summary_df = pd.DataFrame(operations, columns=["#", "Operation", "Key Method"])
summary_df
```
