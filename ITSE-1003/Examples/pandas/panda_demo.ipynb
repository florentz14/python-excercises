# Panda demo (ITSE-1003)

Jupyter version of `panda_demo.py`: dictionary → DataFrame, filters, new columns, and quick analysis.

```python
import pandas as pd
```

## Step 1: build the dataset (dictionary)

```python
data = {
    "name": ["John", "Jane", "Jim", "Jill"],
    "age": [25, 30, 35, 40],
    "major": ["Computer Science", "Mathematics", "Physics", "Chemistry"],
    "gpa": [3.5, 3.8, 4.0, 3.2],
}
```

## Step 2: original data

```python
print("\nOriginal data set:")
print(data)
```

## Steps 3–4: DataFrame

```python
df = pd.DataFrame(data)
print("\nDataFrame:")
df
```

## Question 1: without pandas (age > 20)

```python
print("\nWithout pandas:")
for i in range(len(data["age"])):
    if data["age"][i] > 20:
        print(data["name"][i])
```

## Question 2: with pandas — oldest student(s)

```python
print("\nWith pandas:")
oldest_student = df[df["age"] == df["age"].max()]
oldest_student
```

## Step 7: new column `grade` (= gpa × 10)

```python
df["grade"] = df["gpa"] * 10
print("\nDataFrame with new column:")
df
```

## Step 8: boolean column `senior` (age > 20)

```python
df["senior"] = df["age"] > 20
print("\nDataFrame with new column (senior):")
df
```

## Quick analysis: count by major

```python
print("\nQuick Analysis: Count by major?")
df["major"].value_counts()
```

## Quick analysis: mean GPA by major

```python
print("\nQuick Analysis: Mean GPA by major?")
df.groupby("major")["gpa"].mean()
```
