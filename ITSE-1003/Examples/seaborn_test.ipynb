```python
import seaborn as sns
import pandas as pd
```

```python
sns.scatterplot(x=[1,2,3], y=[1,2,3])
sns.set_theme(style='darkgrid')
sns.color_palette()
```

```python
tips = pd.read_csv('data/tips.csv')
tips
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',hue='sex', kind='scatter', col='day',col_wrap=2, style='smoker')
```

```python
df_down = pd.read_csv("data/dowjones.csv", parse_dates=["Date"])
```

```python
sns.lineplot(data=df_down, x='Date', y='Price')
```

```python
sns.relplot(data=df_down, x='Date', y='Price', kind='scatter')
```

```python
sns.displot(data=df_down, x='Date', y='Price', kind='hist')
```
